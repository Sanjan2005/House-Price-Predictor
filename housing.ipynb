data preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

dataset = pd.read_csv('Housing.csv')
x = dataset.iloc[:, 1:].values
y = dataset.iloc[:, 0].values
```

```python
print(y)
```

```python
print(x)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
cv = ColumnTransformer(transformers=[("encoding",OneHotEncoder(),[5,6,7,8,9,10,11])],remainder="passthrough")

x=np.array(cv.fit_transform(x))
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
y_train = y_train.reshape(len(y_train),1)
print(y_train)
```

```python
x_train = np.delete(x_train,-1,1)
```

```python
print(x_train)
```

```python
x_test = np.delete(x_test,-1,1)
```

```python
print(x_test)
```

```python
y_test = y_test.reshape(len(y_test),1)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
```

Output of model

```python
print(y_pred)
```

Comparision betweeen (Predicted values, Original values)

```python
print(np.c_[y_pred, y_test])
```

```python
from sklearn.metrics import r2_score,accuracy_score
r2_score(y_test,y_pred)
```
